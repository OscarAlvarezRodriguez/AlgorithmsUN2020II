```
#TOMADO DE https://es.wikipedia.org/wiki/Ordenamiento_de_burbuja   ESTE ES SIN OPTIMIZAR

def bubbleSort(array):
  n = len(array)
  pasos = 0;
  for i in range(n):
    for j in range(0, n-1):
      pasos += 1;
      if array[j] > array[j+1]:
        pasos += 1;
        array[j], array[j+1] = array[j+1], array[j]
  return pasos;
```

```
import numpy as np

def bubbleSortOptimized(array):
  n = len(array);
  MyArray = np.array(array);
  flag = True
  pasos = 0;
  while(flag):
    flag = False;
    for i in range(n-1):
      pasos += 1;
      if MyArray[i] > MyArray[i+1]:
        pasos += 1;
        MyArray[i], MyArray[i+1] = MyArray[i+1], MyArray[i]
        flag = True;
  return pasos;
```

```
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors

def GraficarBarra(title, datos):
  if(len(datos) != 1):
    n = max(datos);
    if(n == 0):
      return
    
    array = np.zeros(n);
    arrayx = np.zeros(n);
    for i in range(0,n-1):
      arrayx[i] = i;

    for i in range(0,len(datos)-1):
      array[datos[i]] += 1;

    fig, axs = plt.subplots(1, 1, sharey=True, tight_layout=True)
    plt.title("Grafica para: " + str(title) + " datos")
    axs.bar(arrayx,array)
    return
  fig, axs = plt.subplots(1, 1, sharey=True, tight_layout=True)
  plt.title("Grafica para: " + str(title) + " datos")
  axs.hist(datos, bins=1)

def GraficarHistograma(title, datos):
  fig, axs = plt.subplots(1, 1, sharey=True, tight_layout=True)
  plt.title("Grafica para: " + str(title) + " datos")
  # axs.hist(datos, bins=(max(datos)-min(datos)))
  axs.hist(datos, bins=(round(max(datos)/2)))
```

```
x = [];
for i in range(1,10):
  x.append(i);
  y = show(permutations(x))
  pasos = [];
  for j in range(0,len(y)):
    pasos.append(bubbleSortOptimized(y[j]))
  if (len(x)==1):
    GraficarBarra(i,pasos);
    pasos.clear();
  else:
    GraficarBarra(i,pasos);
    pasos.clear();
#show(permutations(x))
```
